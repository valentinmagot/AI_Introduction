```python
BCWS = 5000 + 11000 + 12000 + 12000
BCWP = 5000 + 11000 + 12000 
ACWP = 4000 + 11000 + 13000

SPI = BCWP/BCWS
CPI = BCWP/ACWP
CR = SPI * CPI

print('La valeur de SPI est : ' + str(SPI))
print('La valeur de CPi est : ' + str(CPI))
print('La valeur de CR est : '+ str(CR))
```

Comme la valeur de SPI est < 1 cela signifie que le projet est en retard sur l'echeancier. Ensuite comme la valeur de CPI est = 1 alors cela signifie que le projet est sur le budget. Ainsi comme la valeur de CR est < 0.8 cela signifie que l'etat generale du projet est problematique.
