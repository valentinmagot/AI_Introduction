Un autre concept important est que les cellules dépendent les unes des autres. Par exemple, si nous définissons une fonction dans une cellule, cette cellule (ou plus précisément la fonction y définie) doit être exécutée avant de pouvoir être appelée. En outre, après avoir modifié une cellule, cette cellule doit être réexécutée pour que la mise à jour prenne effet. Cela peut être vu ci-dessous.

```python
def add(num1, num2):
    '''
    Retourne l'addition de num1 et num2
    '''
    return num1 + num2
```

```python
# Add 1 and 3
print("1 + 3 =", add(1, 3))
```

S'il y a une erreur, elle sera affichée dans la cellule de sortie

```python
def add():
    '''
    Nous aurons une erreur ici puisque num1 et num2 n'existent pas
    '''
    return num1 + num2
```

```python
# On appelle la fonction qui va causer une erreur
print("1 + 3 =", add())
```
